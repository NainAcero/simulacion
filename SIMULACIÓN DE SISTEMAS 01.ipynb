```python
def randint_list(n):
    x = []
    for i in range(n):
      x.append(np.random.random())
    return x
```

```python
datos = randint_list(200)
print(datos)
dato_ln = []
for dato in datos:
    ln = np.log(1-dato)
    dato_ln.append(-3*ln)
```

```python
plt.plot(datos, dato_ln, 'ro')
```

```python
plt.plot(dato_ln)
```
