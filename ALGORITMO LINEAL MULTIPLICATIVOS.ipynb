## ALGORITMO LINEAL MULTIPLICATIVOS  - ESIS ( Nain Acero)

```python
import numpy as np
import pandas  as pd
import os
import matplotlib.pyplot as plt
```

```python
def lineal_multiplicativos(n, semilla, a, m, c):
    x = []
    for i in range(n):
        numero = ((a*semilla) + c) % m
        numero1 = numero/(m-1)
        x.append(numero1)
        semilla = numero
    return x
```

```python
semilla = 6
a = 13
m = 8
c = 7
data = lineal_multiplicativos(50, semilla, a, m, c)
print(data)
df = pd.DataFrame({'column':data}) 
df.to_csv('out_algoritmo_congruensales_multiplicativos.csv', index=False, header=['aleatorio'],sep=';',decimal='.')
df.to_csv('out_algoritmo_congruensales_multiplicativos.txt', index=False, header=['aleatorio'],sep=';',decimal='.')
```
