## ALGORITMO LINEAL MULTIPLICATIVOS  - ESIS ( Nain Acero)

```python
import numpy as np
import pandas  as pd
import os
import matplotlib.pyplot as plt
```

```python
def lineal_multiplicativos(n, semilla, cmultiplicativa, modulo):
    x = []
    for i in range(n):
        numero = (cmultiplicativa*semilla) % modulo
        numero1 = numero/(modulo-1)
        x.append(numero1)
        semilla = numero
    return x
```

```python
semilla = 15
cmultiplicativa = 35
modulo = 64
data = lineal_multiplicativos(50, semilla, cmultiplicativa, modulo)
print(data)
df = pd.DataFrame({'column':data}) 
df.to_csv('out_algoritmo_congruensales_multiplicativos.csv', index=False, header=['aleatorio'],sep=';',decimal='.')
df.to_csv('out_algoritmo_congruensales_multiplicativos.txt', index=False, header=['aleatorio'],sep=';',decimal='.')
```
