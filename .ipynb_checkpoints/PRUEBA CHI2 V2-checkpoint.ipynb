```python
import numpy as np
import pandas  as pd
import os
import matplotlib.pyplot as plot
import statistics as stats
import scipy.stats as st 
```

```python
data = [14, 7, 13, 16, 16, 13, 14, 17, 15, 16,
        13, 15, 10, 15, 16, 14, 12, 17, 14, 12,
        13, 20, 8, 17, 19, 11, 12, 17, 9, 18,
        20, 10, 18, 15, 13, 16, 24, 18, 16, 18,
        12, 14, 20, 15, 10, 13, 21, 23, 15, 18]
```

```python
print(stats.mean(data))
```

```python
print(stats.pvariance(data))
```

```python
intervalos = range(min(data), max(data) + 2)

plot.hist(x=data, bins=intervalos, color='#F2AB6D', rwidth=0.85)
plot.title('Histograma de Números')
plot.xlabel('Edades')
plot.ylabel('Frecuencia')
plot.xticks(intervalos)

plot.show() #dibujamos el histograma
```

```python
datos = []
for val in data:
    datos.append(val/99)
```

```python
def prueba_chi_cuadrada(pseudoaleatorios):
  n = len(pseudoaleatorios)
  m = int(n**0.5)
  intervalo = 1/m
  frec_esperada = np.ones(m)*(n/m)
  frec_observada = np.zeros(m)
  for pseudoaleatorio in pseudoaleatorios:
    posicion = int(pseudoaleatorio/intervalo) if pseudoaleatorio!=1 else m-1
    frec_observada[posicion] += 1
  return (((frec_esperada-frec_observada)**2/frec_esperada).sum(), frec_observada)

pseudoaleatorios = np.array(datos)
print("El estadístico chi-cuadrada es: ",prueba_chi_cuadrada(pseudoaleatorios)[0])
a = 0.05
print("Tabla", st.chi2.ppf( 1 - (a/2), int(len(datos)**0.5) - 1))
x = prueba_chi_cuadrada(pseudoaleatorios)[1].astype(int)
```

```python
val = []
for i in range(len(x)):
    for j in range((x[i])):
        val.append(i+1)
```

```python
intervalos = range(min([1,2,3,4,5,6,7]), max([1,2,3,4,5,6,7]) + 2)
print(x)
plot.hist(x=val, bins=intervalos, color='#F2AB6D', rwidth=0.85)
plot.title('Histograma de Números')
plot.xlabel('Edades')
plot.ylabel('Frecuencia')
plot.xticks(intervalos)

plot.show() #dibujamos el histograma
```
